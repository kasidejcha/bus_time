```python
import pandas as pd
import random
```

```python
df1 = pd.read_csv('/Users/kasidej/Documents/bus_time/archive/Regularities_by_liaisons_Trains_France.csv')
df2 = pd.read_csv('/Users/kasidej/Documents/bus_time/archive/Travel_titles_validations_in_Paris_and_suburbs.csv')
df1.head()
```

```python
df = df1.iloc[:,0:5]
df.Month = df.Month.astype(int)
df['date'] = pd.to_datetime(df.Year.astype(str) + '/' + df.Month.astype(str) + '/01')
df = df.drop(['Year', 'Month'], axis=1)
df
```

```python
mapping = pd.DataFrame(df['Departure station'].value_counts()).reset_index()\
            .drop('Departure station', axis=1).rename(columns={'index':'stations'}).reset_index().set_index('stations')
print(mapping.head())
mapping = mapping.to_dict()['index']
df['depart_station'] = df['Departure station'].map(mapping)
df['arrival_station'] = df['Arrival station'].map(mapping)
df = df.drop(['Departure station', 'Arrival station'], axis=1)
df
```

```python
df19_7 = df[df.date == '2019-07-01']
df19_7
```

```python
# find route
destination = []
for i in range(len(df19_7)):
    if i == 0:
        arrive = random.choice(df19_7.iloc[:]['arrival_station'].tolist())
        destination.append(arrive)
        count = 0

    # find possible arrive station
    possible_arrive = df19_7[df19_7['depart_station']==arrive]
    print(possible_arrive)
    # possible_arrive = possible_arrive[~possible_arrive['arrival_station'].isin(destination)]
    arrive = random.choice(possible_arrive['arrival_station'].tolist())
    print('arrive:',arrive)
    destination.append(arrive)
    print('destination:',destination)
    if len(destination) > 10:
        break
                
```

```python
destination = [17, 0, 4, 6, 3, 0]
```

```python
count = 0
df = df.sort_values('date').reset_index(drop=True)
for date in sorted(set(df['date'].tolist())):
    input_data = df[df.date == date]
    for i in range(len(destination)-1):
        if i == 0:
            data = input_data[(input_data.depart_station==destination[i]) & (input_data.arrival_station == destination[i+1])]
        else:
            data = pd.concat([data,input_data[(input_data.depart_station==destination[i]) & (input_data.arrival_station == destination[i+1])]])
            data['link'] = data['depart_station'].map(str) + ':' + data['arrival_station'].map(str)
    if count == 0:
        total_data = data
        count += 1
    else:
        total_data = pd.concat([total_data, data])
total_data = total_data.drop(['depart_station', 'arrival_station'], axis=1).reset_index(drop=True)
total_data
```
